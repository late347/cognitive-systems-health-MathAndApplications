```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 6)
# Design matrix X = [1 x]
X = np.concatenate( (np.ones((5,1)), x.reshape(5,1), x.reshape(5,1)**2),  axis=1 )
X = np.concatenate((X, x.reshape(5,1)**3, x.reshape(5,1)**4), axis=1)
X
```

```python
xi = np.arange(1,11).reshape(10,1)
yi=np.array([14,12,34,24,43,38,41,57,55,50]).reshape(10,1)

X0= np.concatenate((np.ones( (10,1) ), xi),axis=1)
X0
```

```python
M0=np.linalg.inv(  X0.T.dot(X0) ).dot(X0.T).dot(yi)
M0
```

```python
xx=np.linspace(1,10,1000)
xx
XX= np.concatenate( ( np.ones((1000,1)), xx.reshape(1000,1)  ),axis=1)
yy=XX.dot(M0)
plt.plot(xi,yi,'o')
plt.plot(xx,yy)
plt.show()

#looks like reasonable results to my eye, but the error looks quite large,
#but error looks like its steady
```

```python
X1 = np.concatenate( (np.ones((10,1)), xi , xi**2), axis=1 )
M1= np.linalg.inv( X1.T.dot(X1)  ).dot(X1.T).dot(yi)

xx1 = np.linspace(1,10,1000)
XX1=np.concatenate((np.ones((1000,1)), xx1.reshape(1000,1), xx1.reshape(1000,1)**2), axis=1)

                   
yy1=XX1.dot(M1)        

plt.plot(xx1,yy1)
plt.plot(xi,yi,'o')
plt.show()

#seems to fit slightly better than straight line?!
```

```python

#convert back to 1D array
x2=xi[:,0]
y2=yi[:,0]
print(x2)
#polyfit 1D arrays
coef = np.polyfit(x2,y2,5)
p=np.poly1d(coef)

xx2= np.linspace(1,10,1000)
yy2=p(xx2)
plt.plot(x2,y2,'o')
plt.plot(xx2,yy2)
plt.show()
```

## In my opinion, having looked at the plots

* straight line, or 2nd deg polynomial fits best
* both are quite good, 2nd deg polynomial slightly better
