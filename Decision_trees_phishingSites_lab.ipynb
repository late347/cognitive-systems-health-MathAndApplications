## pre-process the data into input and output for decision tree


```python
# save column headings into a list
colnames = df_ready.columns.get_values()


# extract explanatory variables into a data frame
# takes all input columns except that last output column
X = df_ready.loc[:,'having_IP_Address':'Statistical_report']

# takes the output column only into the Y
# extract response variable (class variable) into a series
Y = df_ready.loc[:, 'Result']
X
```

```python
Y #check result for Y
```

```python
# decision tree classification
classifier = tree.DecisionTreeClassifier(max_depth=3)
classifier.fit(X,Y)
```

```python
# visualize
dot_data = tree.export_graphviz(classifier, out_file=None, feature_names=colnames[:-1], class_names=['legit','PHISHING'])
graph = graphviz.Source(dot_data) 

graph.render("phishing")
IFrame("phishing.pdf", width=900, height=700)
```

```python
# predict
Y_pred = classifier.predict(X)

# output confusion matrix
#first is the groundTruth values, then is the predictions
cm = confusion_matrix(Y, Y_pred)
print("Confusion matrix:\n",cm)

accuracy = (cm[0][0]+cm[1][1])/(cm[0][0]+cm[1][1]+cm[0][1]+cm[1][0])
print("Accuracy calculated from the training set = %.3f" % (accuracy))

print(classification_report(Y, Y_pred, target_names=['legit', 'phishing']))
```
